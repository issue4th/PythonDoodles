```python
import matplotlib.pyplot as plt
import numpy as np

def fn(number):
    return np.tan(number)

def map(func, listOfNumbers):
    result = []
    for number in listOfNumbers:
        result.append(func(number))
    return result

x = np.linspace(-2* np.pi, 2 * np.pi, 1000)

plt.plot(x, map(fn, x))

plt.show()
```
